## Boleanos con arrays

```python
a = np.arange(4)
```

```python
a
```

```python
a[0], a[-1]
```

```python
a[[0, -1]]
```

```python
a[[True, False, False, True]]
```

```python
a >= 2
```

```python
a[a >= 2] #si se cumple la condicion devuelve True
```

```python
a.mean()
```

```python
a[a > a.mean()]
```

```python
a[~(a > a.mean())] #devuelve los objetos que no son mayores a la media
```

```python
a[(a == 0) | (a == 1)] #or
```

```python
a[(a <= 2) & (a % 2 == 0)] #and
```
