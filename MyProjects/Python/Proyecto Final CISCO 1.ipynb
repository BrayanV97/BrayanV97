```python
from random import randrange

def display_board(board):
    # Esta función muestra el tablero de juego en la consola
    print("+-------+-------+-------+")
    for row in range(3):
        print("|       |       |       |")
        print(f"|   {board[row][0]}   |   {board[row][1]}   |   {board[row][2]}   |")
        print("|       |       |       |")
        print("+-------+-------+-------+")

def enter_move(board):
    # La función acepta el estado actual del tablero y pregunta al usuario por su movimiento
    # Verifica la entrada y actualiza el tablero acorde al movimiento del usuario
    while True:
        try:
            move = int(input("Ingresa tu movimiento (1-9): "))
            if move < 1 or move > 9:
                print("¡Número fuera de rango! Debe ser entre 1 y 9.")
                continue
                
            # Convertir el número de casilla a coordenadas de fila y columna
            row = (move - 1) // 3
            col = (move - 1) % 3
            
            # Verificar si la casilla está libre
            if board[row][col] == 'X' or board[row][col] == 'O':
                print("¡Esta casilla ya está ocupada! Elige otra.")
                continue
                
            # Hacer el movimiento
            board[row][col] = 'O'
            break
        except ValueError:
            print("¡Entrada inválida! Debes ingresar un número.")

def make_list_of_free_fields(board):
    # La función examina el tablero y construye una lista de todos los cuadros vacíos
    free_fields = []
    for row in range(3):
        for col in range(3):
            if isinstance(board[row][col], int):
                free_fields.append((row, col))
    return free_fields

def victory_for(board, sign):
    # La función analiza el estado del tablero para verificar si el jugador que usa 'sign' ha ganado
    # Verificar filas
    for row in range(3):
        if board[row][0] == board[row][1] == board[row][2] == sign:
            return True
    
    # Verificar columnas
    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col] == sign:
            return True
    
    # Verificar diagonales
    if board[0][0] == board[1][1] == board[2][2] == sign:
        return True
    if board[0][2] == board[1][1] == board[2][0] == sign:
        return True
    
    return False

def draw_move(board):
    # La función elige aleatoriamente una casilla libre para la máquina
    free_fields = make_list_of_free_fields(board)
    if free_fields:
        row, col = free_fields[randrange(len(free_fields))]
        board[row][col] = 'X'

# Inicializar el tablero
board = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

# La máquina siempre comienza con una X en el centro
board[1][1] = 'X'

# Mostrar el tablero inicial
print("¡Bienvenido al juego de Tic-Tac-Toe!")
print("La máquina juega con 'X' y tú juegas con 'O'")
print("La máquina ha hecho su primer movimiento:")
display_board(board)

# Bucle principal del juego
while True:
    # Turno del usuario
    enter_move(board)
    
    # Verificar si el usuario ha ganado
    if victory_for(board, 'O'):
        display_board(board)
        print("¡Has ganado! ¡Felicidades!")
        break
    
    # Verificar si hay empate
    if len(make_list_of_free_fields(board)) == 0:
        display_board(board)
        print("¡Empate!")
        break
    
    # Turno de la máquina
    draw_move(board)
    display_board(board)
    
    # Verificar si la máquina ha ganado
    if victory_for(board, 'X'):
        print("¡La máquina ha ganado!")
        break
    
    # Verificar si hay empate después del movimiento de la máquina
    if len(make_list_of_free_fields(board)) == 0:
        print("¡Empate!")
        break
```
