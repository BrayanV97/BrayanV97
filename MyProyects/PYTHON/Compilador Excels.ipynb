```python
import os 
import openpyxl 
import pandas as pd
from openpyxl.styles import PatternFill, Border, Side 
from openpyxl.chart import BarChart, Reference 
import matplotlib.pyplot as plt 

# Ruta de la carpeta que contiene los archivos de Excel 
carpeta = r'C:\Users\braya\OneDrive - ESCUELA COLOMBIANA DE INGENIERIA JULIO GARAVITO\ANT\BPMOTROS'
 
# Diccionario para almacenar los resultados consolidados por archivo 
resultados_consolidados = {} 
 
# Obtener una lista de todos los archivos en la carpeta 
archivos_en_carpeta = os.listdir(carpeta) 

# Filtrar los archivos de Excel 
archivos_excel = [archivo for archivo in archivos_en_carpeta if archivo.endswith('.xlsx')] 

# Crear un DataFrame vacío para almacenar los resultados consolidados
BPM_Consolidado = pd.DataFrame()

# Iterar sobre los archivos de Excel 
for archivo_excel in archivos_excel: 
    # Construir la ruta completa del archivo 
    ruta_archivo = os.path.join(carpeta, archivo_excel) 
    
    # Imprimir el nombre del archivo
    print(f"Procesando archivo: {archivo_excel}")
    
    # Cargar el archivo de Excel y abrimos la hoja BPM
    BPM = pd.read_excel(ruta_archivo, sheet_name='BPM', skiprows=3)
    print(BPM.columns)

    
    # Elimina filas con datos nulos en la columna 'QR'
    BPM_proc = BPM.dropna(subset=['QR']) if BPM['QR'].isnull().values.any() else BPM
    
    # Agregar el DataFrame procesado al DataFrame consolidado
    BPM_Consolidado = pd.concat([BPM_Consolidado, BPM_proc])

BPM_Consolidado['CODIGO DANE'] = BPM_Consolidado['QR'].astype(str).str[:5]

# Guardar el DataFrame consolidado en un archivo Excel
BPM_Consolidado.to_excel('BPM_Consolidado.xlsx', index=False)
```
