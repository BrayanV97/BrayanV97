## Cleaning not null values

```python
df = pd.DataFrame({
    'Sex': ['M', 'F', 'F', 'D', '?'],
    'Age': [29, 30, 24, 290, 25],
})
df
```

Los valores de 290 años y sexo D y ? representa valores invalidos

```python
df['Sex'].unique()
```

```python
df['Sex'].value_counts()
```

```python
df['Sex'].replace('D', 'F') #Reemplaza la F por la D
```

```python
df['Sex'].replace({'D': 'F', 'N': 'M'}) #Reemplaza la F por la D y la N por la M
```

```python
df.replace({
    'Sex': {
        'D': 'F',
        'N': 'M'
    },
    'Age': {
        290: 29
    }
})
```

```python
df[df['Age'] > 100]
```

```python
df.loc[df['Age'] > 100, 'Age'] = df.loc[df['Age'] > 100, 'Age'] / 10 
#podemos dividir por 10 ya que pudo ser un error de digitacion y la verdadera edad fuese 29
```

```python
df
```

## Valores Duplicados

```python
ambassadors = pd.Series([
    'France',
    'United Kingdom',
    'United Kingdom',
    'Italy',
    'Germany',
    'Germany',
    'Germany',
], index=[
    'Gérard Araud',
    'Kim Darroch',
    'Peter Westmacott',
    'Armando Varricchio',
    'Peter Wittig',
    'Peter Ammon',
    'Klaus Scharioth '
])
```

```python
ambassadors #Queremos dejar un embajador por pais
```

```python
ambassadors.duplicated() #Muestra los valores duplicados despues de leer el primero
```

```python
ambassadors.duplicated(keep='last') #Muestra los valores duplicados despues de leer el ultimo
```

```python
ambassadors.duplicated(keep=False) #Muestra todos los valores que tengan duplicados
```

```python
ambassadors.drop_duplicates() #Elimina los valores duplicados despues del primero
```

```python
ambassadors.drop_duplicates(keep='last') #Mantiene el ultimo valor que no este duplicado y elimina los demas
```

```python
ambassadors.drop_duplicates(keep=False) #Elimina todos los valores que tengan algun valor duplicado
```
